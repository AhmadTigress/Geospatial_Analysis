```python
import geopandas as gpd
```

```python
full_data = gpd.read_file("zip://DEC_lands.shp.zip")
full_data.head()
```

```python
type(full_data)
```

```python
data = full_data.loc[:, ['CLASS', 'COUNTY', 'geometry']].copy()
```

```python
# How many lands of each type are there
data.CLASS.value_counts()
```

```python
# Select the land that fall under the "WILD FOREST" or "WILDERNESS" category
wild_lands = data.loc[data.CLASS.isin(["WILD FOREST", "WILDERNESS"])].copy()
```

```python
wild_lands.head()
```

```python
#xyz = data.CLASS.isin(["WILD FOREST", "WILDERNESS"]).copy()
xyz = data[["CLASS"]].isin(["WILD FOREST", "WILDERNESS"])
xyz.head()
```

```python
# Visualize on map
wild_lands.plot()
```

```python
# View the first five entries in the 'geometry' column
wild_lands[["geometry"]].head()
```

```python
# Campsite in New York state (points)
POI_data = gpd.read_file("Decptsofinterest.dbf.zip")
```

```python
#POI_data.head()
campsites = POI_data.loc[POI_data.ASSET=='PRIMITIVE CAMPSITE'].copy()
```

```python
# Foot trails in New York state (LineString)
roads_trails = gpd.read_file("Decroadstrails.dbf.zip")
#roads_trails.head()
trails = roads_trails.loc[roads_trails.ASSET=='FOOT TRAILS'].copy()
```

```python
# County boundries in New York states (Polygon)
counties = gpd.read_file("NY_county_boundaries.dbf.zip")
#counties.head()
```

**Visualize on map**

```python
# Define a base map with county boundaries
ax = counties.plot(figsize=(10, 10), color='none', edgecolor='gainsboro', zorder=3)

# Add 'wild_lands', 'campsites' and 'foot trails' to the base map
wild_lands.plot(color='lightgreen', ax=ax)
campsites.plot(color='maroon', markersize=2, ax=ax)
trails.plot(color='black', markersize=1, ax=ax)
```
